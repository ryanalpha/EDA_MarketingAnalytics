* Add Age and Generation 

```python
df['Age'] = 2022 - df['Year_Birth']
```

```python
def category(Age):
    if Age >= 25 and Age <= 40:
        x = 'Millenial'
    elif Age >= 41 and Age <= 56:
        x = 'X'
    else:
        x = 'Baby Boomer'
        
    return x
```

```python
df['Generation'] = df['Age'].apply(lambda x: category(x))
```

## 1. Bagaimana tingkat kepuasan Customer

```python
df_customer = (pd.crosstab(index=df['Complain'], columns='Percentage', normalize=True) * 100).round(0).reset_index()
```

```python
df_customer 
```

```python
plt.figure(figsize=(8,7))
sns.barplot(x='Complain', y='Percentage', data=df_customer, hue='Percentage', palette=['red','lawngreen'])
plt.title("Customer satisfication (%)")
plt.savefig("Satisfication.png")
```

### Tingkat kepuasan customer tersebut mencapai 99% https://www.proprofsdesk.com/blog/tips-for-reducing-customer-complaints/

## 2. Bagaimana persebaran tiap customer 

```python
df_generation = (pd.crosstab(index=df['Generation'], columns='Percentage', normalize=True) * 100).round(0).reset_index()
df_generation
```

```python
plt.figure(figsize=(8,7))
splot = sns.barplot(df_generation['Generation'], df_generation['Percentage'], palette='inferno')
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 7), 
                   textcoords = 'offset points')

plt.title("Distribution Customer according Generation (%)")
plt.savefig("Generation.png")
```

### Generasi X merupakan customer yang terbanyak mencapai 46% https://www.mentalfloss.com/article/609811/age-ranges-millennials-and-generation-z

## 3. Berapa kunjungan customer ke website setiap bulan

```python
df_visit = pd.crosstab(index='Web Visit', columns=df['Generation'], values=df['NumWebVisitsMonth'], aggfunc='sum').round(0).reset_index()
df_visit1 = df_visit.rename(columns={'row_0' : ''}).set_index('').T.reset_index()
```

```python
plt.figure(figsize=(8,7))
splot = sns.barplot(df_visit1['Generation'], df_visit1['Web Visit'], palette='hot')  
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 7), 
                   textcoords = 'offset points')
plt.title("Average Web Visit each Month", size=14)
plt.savefig("Web.png")
```

### customer mengakses website 1800 - 5801x setiap bulan https://www.searchenginejournal.com/over-50-of-local-business-websites-receive-less-than-500-visits-per-month/338137/

## 5. Bagaimana perkembangan user untuk register ditahun 2013

```python
df_register = pd.crosstab(index=[df['Year'], df['Month']], columns="User", values=df['Month'], aggfunc='count').reset_index()
```

```python
df_register1 = df_register.iloc[[10,9,13,6,14,12,11,7,17,16,15,8]]
df_register1
```

```python
plt.figure(figsize=(15,15))
plt.plot(df_register1['Month'], df_register1['User'], color='darkorange')
plt.xlabel("Month")
plt.ylabel("User")
plt.title("Register in 2013", fontsize=14)
plt.savefig("Register.png")
```
